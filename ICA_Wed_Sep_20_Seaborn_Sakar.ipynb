```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
wisc_data = pd.read_csv('data.csv')
wisc_data
```

```python
pd.set_option('display.max_columns', None)
wisc_data.head()
```

```python
wisc_data.describe()
```

```python
# Investigating NaNs
wisc_data.isna() # Unnamed: 32 has missing values
wisc_data = wisc_data.drop(['Unnamed: 32'], axis=1) # Dropping the column

wisc_data.isna().sum()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=wisc_data, x="diagnosis", y="radius_worst", hue="diagnosis")
plt.title("Comparing the Worst Radius of Tumors by Diagnosis")
plt.xlabel("Diagnosis")
plt.ylabel("Worst Radius")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.kdeplot(data=wisc_data, x="radius_worst", y="concavity_worst", hue="diagnosis")
plt.title("Comparing Radius and Concavity")
plt.xlabel("Radius")
plt.ylabel("Concavity")
plt.show()
```

```python
fg = sns.FacetGrid(wisc_data, col="diagnosis", hue="diagnosis")
fg.map(sns.scatterplot, "smoothness_worst", "radius_worst", alpha=0.5)
fg.add_legend()
```

```python
fg = sns.FacetGrid(wisc_data, col="diagnosis", hue="diagnosis")
fg.map(sns.scatterplot, "concavity_worst", "radius_worst", alpha=0.5)
fg.add_legend()
```
